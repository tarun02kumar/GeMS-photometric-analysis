```python
import os
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import matplotlib.ticker as ticker
```

```python

# Parameters
img_folder = "D:/VM_SHARED_UPDATED_29032021/VM_SHARED/shared_vmware/Paolo/ngc5904_13a/obj/Ks/exp160s/backup"
list_file = 'rtobjKst160.lst'
img_save = True

# Science parameters (fixed)
fov_side = 82.5   # Field of view [degrees]
telescope_reference = [0, 5.4 ]  # Position of the detector reference point [degrees]
square_size = 82.5 / 3600


xtick_locs = [ 229.63, 229.64, 229.65]
xtick_labels = ['229.63', '229.64', '229.65']
ytick_locs = [2.07, 2.08, 2.09]
ytick_labels = ['2.07', '2.08', '2.09']

# Read the list of image filenames
with open(os.path.join(img_folder, list_file), 'r') as f:
    image_filenames = f.read().splitlines()

# Extract header data from FITS files
ra_values = []
dec_values = []
offsets = []

for filename in image_filenames:
    with fits.open(os.path.join(img_folder, filename)) as hdul:
        header = hdul[0].header
        ra_values.append(header['RA'])
        dec_values.append(header['DEC'])
        offsets.append((header['XOFFSET'], header['YOFFSET']))

# Plotting
fig, ax = plt.subplots( figsize=(6,6), dpi= 300)

for i in range(len(ra_values)):
    ra = ra_values[i]
    dec = dec_values[i]
    x_offset, y_offset = offsets[i]
    ax.plot([ra + x_offset/3600], [dec + y_offset/3600], color= 'blue', marker = 'o', ms = 5, linestyle = None)
    rect = Rectangle((ra + x_offset/3600 - square_size/2, dec + y_offset/3600 - square_size/2), 
                     square_size, square_size, fill=False, edgecolor='blue', linestyle ='solid', linewidth=1)
    ax.add_patch(rect)

ax.set_xticks(xtick_locs, xtick_labels, fontsize= 16)
ax.set_yticks(ytick_locs, ytick_labels, fontsize= 16)
ax.tick_params(axis='x', bottom=True, top=True, direction='in', which ='both', labelsize=16)  
ax.tick_params(axis='y', left=True, right=True,  direction='in', which ='both', labelsize=16)  

ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(10))
ax.yaxis.set_minor_locator(ticker.AutoMinorLocator(10))



plt.xlabel('R.A. (deg)', fontsize = 16)
plt.ylabel('Dec (deg)', fontsize = 16)
plt.tight_layout()
if img_save:
    plt.savefig(os.path.join(img_folder, 'dithering_objKs.png'))
plt.show()

print('Program terminated')
print(ra_values)
print(dec_values)
print(offsets)
```

```python


# Size of the square in degrees


# Plotting with squares
plt.figure(figsize=(10, 6))
for i in range(len(ra_values)):
    ra = ra_values[i]
    dec = dec_values[i]
    x_offset, y_offset = offsets[i]
    
    # Plot the point
    plt.plot(ra + x_offset/3600, dec + y_offset/3600, color='red', marker='o', ms=5, linestyle=None)
    
    # Add the square centered on the point
    rect = Rectangle((ra + x_offset/3600 - square_size/2, dec + y_offset/3600 - square_size/2), 
                     square_size, square_size, fill=False, edgecolor='blue', linewidth=1)
    plt.gca().add_patch(rect)

plt.xlabel('R.A. (deg)')
plt.ylabel('Dec (deg)')
plt.tight_layout()

# Save the figure if the flag is set
if img_save:
    plt.savefig(os.path.join(img_folder, 'dithering_with_squares.png'))

plt.show()

print('Program terminated')
print(ra_values)
print(dec_values)
print(offsets)
```
